```python
#Data Extraction
import pandas as pd
names_hd = ["Username","Mac_Address","Time_Duration","Date", "Location","IP_Address", "XYZ","Activity","Service1", "service2","Success"]
my_data = pd.read_csv("data.csv",names=names_hd)
```

```python
#creating a bar graph 
#all locations accessed

import matplotlib.pyplot as plt; plt.rcdefaults()
import numpy as np
import matplotlib.pyplot as plt

objects=[]
objects = set(my_data['Location'])
y_pos = np.arange(len(objects))
performance=my_data['Location'].value_counts()
 
plt.bar(y_pos, performance, align='center', alpha=0.3)
plt.xticks(y_pos, objects)
plt.ylabel('Count')
plt.title('Locations Accessed')
 
plt.show()
```

```python
# Bar graph of top 10 locations accessed.

obj=[]
obj = set(my_data['Location'])
y_pos = np.arange(len(obj))
performance=dict(my_data['Location'].value_counts())
nSites = sorted(performance.values(), reverse = True)
nSites=nSites[:10]
lands = sorted(performance.keys(), key=performance.__getitem__, reverse=True)
lands=lands[:10]
print ("\nMost Frequent Locations Accessed are\n",lands)
import matplotlib.pyplot as pypl
landAsX = range (10)
pypl.bar (lands, nSites)
pypl.grid(True, axis='y')
pypl.show()
#plt.show()g                                                    
```
